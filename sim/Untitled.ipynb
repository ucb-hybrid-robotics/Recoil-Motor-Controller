```python
import numpy as np
from matplotlib import pyplot as plt

from matplotlib import animation
from IPython.display import HTML

plt.rcParams['figure.figsize'] = [16, 10]
```

```python

class SimMotor:
    def __init__(self):
        self.position = 0
        self.velocity = 0
        self.acceleration = 0
        self.inertia = 1
    
    def update(self, dt, torque):
        self.acceleration = torque / self.inertia
        self.velocity += self.acceleration * dt
        self.position += self.velocity * dt

class Environment:
    def __init__(self):
        self.motor = SimMotor()
        self.dt = 1 / 1e3
        self.t = 0
        self.observation = {}
        
    def step(self, inputs):
        self.motor.update(self.dt, inputs["motor_torque"])
        self.t += self.dt
        
        self.observation["motor_position"] = self.motor.position
        self.observation["motor_velocity"] = self.motor.velocity
        self.observation["motor_acceleration"] = self.motor.acceleration

    def getTimeHistory(self):
        self.t_arr = np.linspace(0, self.t, int(self.t // self.dt))
        return self.t_arr
        
env = Environment()

inputs = {
    "motor_motor_torque": 0.1
}

pos_arr = []

for i in range(100):
    if i < 50:
        inputs["motor_torque"] = 0
    else:
        inputs["motor_torque"] = 1
    env.step(inputs)

    print(env.observation["motor_position"])
    pos_arr.append(env.observation["motor_position"])

t_arr = env.getTimeHistory()

```

```python
plt.plot(t_arr, pos_arr, label="Position")
plt.legend(loc="upper right")
plt.title("Motion History")
```
